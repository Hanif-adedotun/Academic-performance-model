```python
#Decision Tree Algorithm
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = pd.read_csv('student-por.csv')

data['school'] = data['school'].map(
                   {'GP':1 ,'MS':0})
data['sex'] = data['sex'].map(
                 {'M':1 ,'F':0})         
data['address'] = data['address'].map(
                   {'U':1 ,'R':0})  

data['famsize'] = data['famsize'].map(
                   {'GT3':1 ,'LE3':0})  
data['Pstatus'] = data['Pstatus'].map(
                   {'A':1 ,'T':0})  
 
data['reason'] = data['reason'].map(
                   {'course':3 ,'home':2,'reputation':1,'other':0})  
data['guardian'] = data['guardian'].map(
                   {'father':1 ,'mother':0})  


data.shape
model = DecisionTreeClassifier()

input = data.drop(columns=['G3','traveltime','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic','famrel','freetime','goout','Dalc','Walc','health','absences','Fjob','Mjob','Medu','address','reason','guardian','failures','school','famsize','Pstatus','Fedu'])
output = data['G3']
input_train,input_test,output_train,output_test = train_test_split(input,output,test_size=0.2)
model.fit(input_train,output_train)
prediction = model.predict(input_test)
accuracy = accuracy_score(output_test,prediction)

#Results
accuracy = accuracy * 100
print("accuracy:")
print(accuracy)

print("output_test:")
print(output_test)

print("prediction:")
print(prediction)

```
